4个输入单元，3个输出单元

如果用$\pmb X$表示输入，$\pmb O$表示输出， $\pmb H$表示隐藏层。

$$\pmb H = \pmb X · \pmb w _1+ \pmb b_1 $$
$$\pmb O = \pmb H · \pmb w _2+ \pmb b_2 $$

在大多数情况下，线性模型不能很好的拟合模型，这时候将Hidden layer的拟合关系变成曲线，就可以很好的适合不同的场景，这个曲线Function叫做Activation Function。

## Activation Function
- 1. Sigmoid Function

$$ sigmoid(x) = \frac{1}{1+e^{-x}}$$

```python
import torch
from d2l import torch as d2l
```

```python
def sigmoid(x):
    return 1/(1+torch.exp(-x))
```

```python
x = torch.arange(-5,5,0.01)
y = sigmoid(x)
```

```python
d2l.plot(x.detach(), y.detach(), 'x', 
         'sigmodi(x)', figsize=(5, 2.5),
        )
```

- 2. Tanh Function

$$ tanh(x) = \frac{1-e^{-2x}}{1+e^{-2x}}$$

```python
def tanh(x):
    return (1-torch.exp(-2*x))/(1+torch.exp(-2*x))
```

```python
x = torch.arange(-5,5,0.01)
y = tanh(x)
```

```python
d2l.plot(x.detach(), y.detach(), 'x', 
         'sigmodi(x)', figsize=(5, 2.5),
        )
```

- 3. ReLU Function

$$ ReLU(x) = max(x,0)$$

```python
x = torch.arange(-5.0, 5.0, 0.01, requires_grad=True)
y = torch.relu(x)
d2l.plot(x.detach(), y.detach(), 'x', 'relu(x)', figsize=(5, 2.5))
```

relu function 突出的优势是求导简单且不存在gradient explosion的问题。

```python
y.backward(torch.ones_like(x), retain_graph=True) 
d2l.plot(x.detach(), x.grad, 'x', 'grad of relu', figsize=(5, 2.5))
```
